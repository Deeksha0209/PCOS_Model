# **Cleaning the Dataset- involves removing unnecessary columnms, unnamed values, replacingmissing values or dropping duplicate values.**

```python
##Dropping columns which we dont require in our prediction
df = df.drop(["Sl. No","Patient File No."],axis = 1)
```

```python
#Removing unnamed columns from df
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
#let's take a look at the AMH column. AMH is Anti-Muller Hormone related to follicles in ovary.
df["AMH(ng/mL)"].head()
```

```python
#Let's take a look at another object dtype:  beta-HCG
df["II    beta-HCG(mIU/mL)"].head()
```

```python
#Converting object type to numerical

df["AMH(ng/mL)"] = pd.to_numeric(df["AMH(ng/mL)"], errors='coerce')
df["II    beta-HCG(mIU/mL)"] = pd.to_numeric(df["II    beta-HCG(mIU/mL)"], errors='coerce')
```

# **Starting with Univariate Analysis to draw insights for individual features:**

```python
# Let's create a Bar graph for categorical values

colors = ['green', 'red']
```

```python
#Definig a function bar_plot for cumulative grouping and better visualization
def bar_plot(variable):

    #get feature
    var = df[variable]

    #count number of categorical variable
    Count = var.value_counts()

    # plotting
    plt.figure(figsize = (9,3))
    plt.bar(Count.index,Count,color=colors)
    plt.xticks(Count.index,Count.index.values)
    plt.ylabel("Count")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,Count))
```

```python
#Let's group all the categories we want to consider for our bar plots & visualise the count for each category individually
category = ["Pregnant(Y/N)", "Weight gain(Y/N)", "Skin darkening (Y/N)", "Hair loss(Y/N)",
            "Pimples(Y/N)", "Fast food (Y/N)", "Reg.Exercise(Y/N)"]

for c in category:
    bar_plot(c)
```

## Moving on, now we have to work on using Machine Learning Algos to determine performance metrices suchas accuracy and prediction of our data. We do this by splitting our dataset into train and test data from sklearn library followed by importing various regression/classification algorithms.

##Test-Train split from sklearn

```python
X = df.drop(columns = ["PCOS (Y/N)"])
y = df["PCOS (Y/N)"]
```

```python
#Let's display top 5 entries of our dataset!
X.head()
```

```python
#importing libraraies
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#test size = 0.2 signifies 20% data will be set aside for testing and 80% data for training
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
#now lets import regression and classification alogos from skleanr lib
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
#Setting the random state for our model.
random_state = 25
classifier = [RandomForestClassifier(random_state = random_state),
             LogisticRegression(random_state = random_state)]
```

```python
#herein i determine the paremetrs for my RF classifier.
random_forest = {"max_features": ['auto', 'sqrt'],
                "n_estimators":[300,500,700],
                "criterion":["gini"],
                'max_depth' : [4,5,6,7,8,9,10]}
```

```python
#similarly, setting parameters fro my Logistic regression model
logistic_regression = {"C":np.logspace(-3,3,7),
                    "penalty": ["l1","l2"]}
```

```python
classifier_grid = [random_forest,
                   logistic_regression]
```

```python
cv_result = []
best_estimators = []
```

```python
# fitting model to find best accuracy out of the 2 models

for i in range(len(classifier)):
    clf = GridSearchCV(classifier[i], param_grid = classifier_grid[i], cv = StratifiedKFold(n_splits = 20), scoring = "accuracy", n_jobs = -1, verbose = 1)
    clf.fit(X_train,y_train)
    cv_result.append(round(clf.best_score_*100, 2))
    best_estimators.append(clf.best_estimator_)
    print(cv_result[i])
```

```python
# selecting the best estimators used in hyperparameter tuning

best_estimators
```

```python
#I am setting the variables rf & lr with the required index for best estimator for my models.
rf = best_estimators[0]
lr = best_estimators[1]
```

##Random Forest Classifier

```python
#fitting the rf model for training data
rf.fit(X_train, y_train)
```

```python
#RF algo will try and predict testing data
y_pred_rf = rf.predict(X_test)
```

```python
#importing the confusion matrix to showcase our performace metrices
from sklearn.metrics import confusion_matrix, recall_score, f1_score, precision_score, accuracy_score
```

```python
#Setting our performance metrices for random forest model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
```

```python
#Voila! Let's print the accuracy, recall, f1 score and precision values as attained by the Rf classifier on our model
print("Random Forest Model Accuracy: ", round(accuracy_rf*100, 2), '%')
print("Random Forest Model Recall:", round(recall_rf*100, 2), '%')
print("Random Forest Model F1 Score:", round(f1_rf*100, 2), '%')
print("Random Forest Model Precision:",  round(precision_rf*100, 2), '%')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
CM_rf = confusion_matrix(y_test, y_pred_rf)
#heatmap is ideal for a confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(CM_rf, annot=True, fmt='d', cmap='viridis')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
plt.show()
```

## Logistic Regression Classifier

```python
#now lets move to another algorithm - LR
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
#calculating our performance metrices for Logistic regression algo
accuracy_lr = accuracy_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
```

```python
print("Logistic Regression Model Accuracy: ", round(accuracy_lr*100, 2), '%')
print("Logistic Regression Model Recall:", round(recall_lr*100, 2), '%')
print("Logistic Regression Model F1 Score:", round(f1_lr*100, 2), '%')
print("Logistic Regression Model Precision:",  round(precision_lr*100, 2), '%')
```

```python
CM_lr = confusion_matrix(y_test, y_pred_lr)
# Create a heatmap of the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(CM_lr, annot=True, fmt='d', cmap='flare')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Lets creat a final bar plot showcasing the performace metrices for both the algos used - Random forest & Logistic Regression!
#this will help us compare and decide which algo is best fit for our model.
plt.figure(figsize=(12, 8))
models = ['Logistic Regression', 'Random Forest']
plt.suptitle("Comparison of Logistic Regression and Random Forest Performance Metrics".upper(), weight='bold', fontsize=14)


plt.subplot(2, 2, 1)
accuracy_scores = [accuracy_lr, accuracy_rf]
plt.bar(models, accuracy_scores, color=['skyblue', 'lightcoral'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Comparison of Logistic Regression and Random Forest - Accuracy", fontsize=10)

plt.subplot(2, 2, 2)
recall_scores = [recall_lr, recall_rf]
plt.bar(models, recall_scores, color=['red', 'lightgreen'])
plt.xlabel("Models")
plt.ylabel("Recall")
plt.title("Comparison of Logistic Regression and Random Forest - Recall", fontsize=10)

plt.subplot(2, 2, 3)
f1_score = [f1_lr, f1_rf]
plt.bar(models, f1_score, color=['darkgreen', 'blue'])
plt.xlabel("Models")
plt.ylabel("F1 Score")
plt.title("Comparison of Logistic Regression and Random Forest - F1 Score", fontsize=10)

plt.subplot(2, 2, 4)
precision_score = [precision_lr, precision_rf]
plt.bar(models, precision_score, color=['magenta', 'coral'])
plt.xlabel("Models")
plt.ylabel("Precision")
plt.title("Comparison of Logistic Regression and Random Forest - Precision", fontsize=10)

plt.tight_layout()
plt.show()

#as visualized in this comparison, random forest classifier seems to be an ideal choice of fit for our PCOS model!
```
